Quiero calcular los acumulados de areas para mi rango de años del censo.

```python
#Genero una lista con los años entre 1982 (el primer registro de area marina) y 2026 y les otorgo valor de 0 
anos_faltan = list(range(1982, 2026))
valores1 = [0]*44
nombre = ['calculo']*44

#Diccionario con los nombres de las columnas.
anos_faltan_hectareas = {
    'year': anos_faltan, 
    'area marina':valores1,
    'DESIG_CAT_ENG':nombre
}

#Genero el DF
valores_faltan_hectareas = pd.DataFrame(anos_faltan_hectareas)
valores_faltan_hectareas
```

```python
#Para saber cuales años tengo en el DF areas marinas 
year_limpio = areas_filtrado['year'].unique()

year_limpio
```

```python
# Selecciono aquellos años que no están en el DF de áreas marinas, es decir los años que faltan 
valores_faltan_hectareas2 = valores_faltan_hectareas[~valores_faltan_hectareas['year'].isin(year_limpio)]
valores_faltan_hectareas2
```

```python
print(areas_filtrado.columns)
```

```python
areas_limpio = areas_filtrado.groupby(['year', 'DESIG_CAT_ENG']).agg({
    'area marina': 'sum'
    }).reset_index()

areas_limpio
```

```python
areas_limpio2 = pd.concat([areas_limpio, valores_faltan_hectareas2], axis=0, ignore_index=True)
areas_limpio3 = areas_limpio2.sort_values(by = 'year').reset_index()
areas_limpio3.drop(['index'], axis=1, inplace=True)

areas_limpio3
```

```python
#Cambia la estructura del DF, cada tipo de área es una columna
areas_pivot = areas_limpio3.pivot_table(
    index='year',
    columns='DESIG_CAT_ENG',
    values='area marina',
    aggfunc='sum' 
)

areas_pivot_filled = areas_pivot.fillna(0)

# Suma acumulativa de todas las areas marinas por tipo de area
areas_acumulativo = areas_pivot_filled.cumsum().reset_index()

areas_acumulativo
```
