```python
# Para convertir las fechas, necesito crear una función que me genere las fechas en formato Día/Mes/Año
def convertidor_fecha(row):
    ano = int(row['year'])
    dia_ano = int(row['yday'])
    # Crear una fecha base para el primer día del año.
    fecha_base = datetime(ano, 1, 1)
    # Añadir el número de días para obtener la fecha correcta
    fecha_resultante = fecha_base + timedelta(days=dia_ano - 1)
    return fecha_resultante.strftime('%d/%m/%Y')

cens['fecha'] = cens.apply(convertidor_fecha, axis= 1)

cens['fecha_formateada'] = pd.to_datetime(cens['fecha'], format='%d/%m/%Y')

cens['month'] = pd.to_datetime(cens['fecha_formateada']).dt.month.astype(str).str.zfill(2)
cens['month'] = cens['month'].astype(int)

cens
```

### Generar columna de nom del mes 

```python
meses = {
    1:'Enero', 
    2:'Febrero',
    3:'Marzo',
    4:'Abril',
    5:'Mayo',
    6:'Junio',
    7:'Julio',
    8:'Agosto',
    9:'Septiembre',
    10:'Octubre',
    11:'Nobiembre',
    12:'Diciembre'
    }
cens['nombre_mes'] = cens['month'].map(meses)
cens
```

### Eliminación de fechas duplicadas

```python
# Para buscar aquellas fechas que están repetidas.
censo_count = cens.groupby(['year','month','yday', 'fecha_formateada']).agg({
    'speciesname':'count'
}).reset_index()
censo_count
```

```python
# Defino aquellas fechas que quiero eliminar. Por cada mes me he quedado con los datos del primer censo del mes.
fechas_eliminar = ('16/03/2007', '15/04/2007', '19/07/2019', '08/04/2023', '05/11/2023', '06/12/2023', '08/12/2023', '24/02/2024', '10/05/2024', '11/05/2024', '06/07/2024', '20/07/2024', '11/08/2024', '09/09/2024', '14/12/2024', '16/03/2025', '10/04/2025','14/04/2025', '27/04/2025','30/04/2025')

# Eliminación de las filas que tengan la columna fecha dentro de la lista que he creado fechas_eliminar
cens_limpio = cens[~cens['fecha'].isin(fechas_eliminar)]

cens_limpio
```

### Detección i eliminación de especies poco comunes o que no sean aves marinas 

```python
# Cambiar el nombre del Cormorán grande.
cens_limpio['speciesname'] = cens_limpio['speciesname'].replace('Cormorant (p. c. carbo)', 'Cormorant')
cens_limpio['speciesid'] = cens_limpio['speciesid'].replace(502, 20)

# Calculo cuales son las especies totales del censo de aves marinas.
especies_totales = cens_limpio['speciesname'].value_counts()

# Calculamos cuales serán las especies raras, que acabaré eliminando del censo.
especies_raras = especies_totales[especies_totales < 20].index
cens_limpio = cens_limpio[~cens_limpio['speciesname'].isin(especies_raras)]

# Eliminar aquellas especies comunes que no son aves marinas 
especies_eliminar = ['Common Sandpiper','Serin', 'Fin Whale', 'Alpine Swift', 'Blue-headed Wagtail', 'Marsh Harrier', 'Striped Dolphin', 'Ocean Sunfish', 'Bottle-Nosed Dolphin', 'Grey Heron', 'Meadow Pipit', 'Little Egret', 'Swallow', 'White Wagtail', 'Peregrine', 'Pallid Swift', 'Crag Martin', 'House Martin', 'Swift']
cens_aus = cens_limpio[~cens_limpio['speciesname'].isin(especies_eliminar)]
cens_aus
```

# Merge de los DFs

```python
# Con los datos del estatus de las aves de Cataluña
censo_status = cens_aus.merge(df_ocells, on = 'Scientific name', how = 'left')

# Con df de temperaturas
censo_status_temp = censo_status.merge(temp_total, how='inner', left_on=['year', 'nombre_mes'], right_on=['year', 'nombre_mes'])

#Cambio el nombre de algunas columnas.
censo_status_temp.rename(columns= {
    'Common name':'Catalan name',
    'Categoria':'Protection status in Catlunya',
    'average_temperature':'Average SST month', 
    'media_hist_temp':'Average SST 1975-2000',
    'diferencia_temp': 'SST change'
},inplace=True)

censo_status_temp['Average SST 1975-2000']= censo_status_temp['Average SST 1975-2000'].round(2)

censo_status_temp.drop(['Classe'],axis=1, inplace=True)

censo_total = censo_status_temp.merge(pesquerias2, on = 'year', how = 'left')

censo_total
```

# Acabo de rellenar algunos missing values

```python
# 1 Hacer un diccionario anidado por especie. 
especies_anidado = {
    'Phalacrocorax carbo': {
        'Ordre':'Pelecaniformes',
        'Catalan name':'Corb marí gros',
        'Protection status in Catlunya':'No Evaluated'},
    
    'Larus ridibunus': {
        'Ordre':'Charadriiformes',
        'Catalan name':'Gavina riallera',
        'Protection status in Catlunya':'No Evaluated'}, 
    
    'Puffinus sp.':{
        'Ordre':'Procel·lariformes',
        'Catalan name':'Baldriga balear/mediterranea',
        'Protection status in Catlunya':'Critically Endangered/Protected'},
    
    'Larus michahellis':{
        'Ordre':'Charadriiformes',
        'Catalan name':'Gavià argentat',
        'Protection status in Catlunya':'No Evaluated'}, 
    
    'Larus fuscus': {
        'Ordre':'Charadriiformes',
        'Catalan name':'Gavià fosc',
        'Protection status in Catlunya':'No Evaluated'},
    }

for scientific_name, values_to_fill in especies_anidado.items():
    mask=censo_total['Scientific name'] == scientific_name
    
    for column, value in values_to_fill.items():
        censo_total.loc[mask, column] = censo_total.loc[mask, column].fillna(value)


censo_total      
```

```python
censo_total['Ordre'] = censo_total['Ordre'].replace('Caradriformes', 'Charadriiformes')
censo_total['Ordre'] = censo_total['Ordre'].replace('Pelecaniformes', 'Suliformes')

censo_total
```
